# Pie Plot

```python
activities = ['sleep', 'read', 'sports', 'eat', 'tv','gossip']
hours = [8,5,2,1,2,3]
col = ['navy','red','blue','green','yellow','violet']
```

```python
plt.pie(hours, labels = activities, colors = col, autopct = '%1.0f%%',
       shadow=True, explode = [0,0.3,0,0,0,0])
plt.figure(figsize = (2,2))
plt.subplot(1,2,2)
plt.scatter(x1, y1, s=100, color = 'blue', marker='s' )
plt.show()
```

# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
churn = pd.read_csv(r"D:\CTTC files\Datasets for practice\customer_churn_dataset-training-master.csv" )
churn
```

```python
churn = churn.set_index('CustomerID')
churn
```

```python
sns.kdeplot(churn.Age)
plt.grid()
plt.show()
```

```python
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
sns.distplot(churn.Age) # histogram + kde
plt.show()
```

```python
sns.displot(churn.Age) # bar graph + kde
plt.show()
```

```python
sns.displot(churn.Age[churn.Gender=='Female']) 
plt.title("Age and Gender")
plt.show()
```

```python
sns.displot(churn.Age[churn['Subscription Type']=='Standard'],color='Blue') 
plt.title("Age and Subscription Type")
plt.show()
```

```python
sns.countplot(x = churn['Subscription Type'] )
plt.title("Subscription Type Count plot")
plt.show()
```

```python
sns.countplot(x = churn['Subscription Type'], hue = churn['Contract Length'] )
plt.title("Subscription Type and Contract length Count plot")
plt.show()
```

# BOX Plot

```python
sns.boxplot( x = churn.Gender, y = churn.Age)
plt.show()
```

```python
sns.boxplot( x = churn.Gender, y = churn['Total Spend'] )
plt.show()
```

```python
sns.boxplot( x = churn.Gender, y = churn['Total Spend'], hue = churn['Contract Length'] )
plt.show()
```

```python
sns.violinplot( x = churn.Gender, y = churn['Total Spend'], hue = churn['Contract Length'] )
plt.show()
```

# Heat Map

```python
churn1 = churn.drop(['Gender','Subscription Type','Contract Length'], axis = 1)
churn1
```

```python
cor = churn1.corr()
cor
```

```python
sns.heatmap(cor, annot=True, cmap = 'magma')
plt.figure(figsize = (12,12))
plt.show()
```
